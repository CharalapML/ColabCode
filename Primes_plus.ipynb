# This first bit simply calculates all primes up to a given number (e.g all primes up to and including, say 111)

```python
import numpy as np
import matplotlib.pyplot as mpl
import time
import math
```

```python

primex=[]
def prime(n):
    '''
    Input:
    n: positive integer. This function checks primes upto n    '''

    if n<0:
        raise ValueError("n should be positive")

    for x in range(2, n):
        if x>1: # Takes all positive numbers including even numbers
            for i in range (2,x):
                if (x%i) == 0:  #Â this divides all numbers up to n by i. If this is a whole number stop this
                                # if a fraction is returned then the number x is not divisible by a whole number up to i
                                # and so is not a prime number
                    break
            else:
                primex.append(x)   # appends x to the file primex
    print(primex)
    print (" Number of primes up to " ,n, "is: ", len(primex))
```

```python
#prime?
prime (111)
```
