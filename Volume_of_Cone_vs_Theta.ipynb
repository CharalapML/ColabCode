# **This program show the variation in the volume of a cone that is created from a disc (circle) with an angle Theta removed to make the cone**

## **The angle Theta removed can be as little as fractionally greater than 0 radians and the maximum fractionally less than 2$\pi$ radians**




V( ) = 1/3 * pi * R^3 * {1 - Y)^2 * {SQRT(2y - y^2)}^0.5

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
print(math.pi)
print ("The value of radians(360) : ",  math.radians(360), "Radians")
```

```python
angleRads = []
angleDegree = []
volume = []
R = 1
pi = math.pi
for theta in range (0, 360, 1):
  rads = math.radians(theta)
  y = rads/(2*pi)
  v =  (1/3)*pi*(R**3)*((1-y)**2)*((2*y-(y**2))**0.5)
  angleRads.append(rads)
  angleDegree.append(theta)
  volume.append(v)

fig,ax = plt.subplots(2)
ax[0].plot(angleRads,volume)
ax[1].plot(angleDegree,volume)
#ax.set_yticks([0,1,2,3])
#ax.set_yticks([0,1,2,3])

Vmax =np.max(volume)
print('For a Radius, where R=1, Max Volume = ', Vmax, 'in degrees this is at angle of ', np.argmax(volume) )


fig,ax = plt.subplots(2)
ax[0].plot(angleRads,volume)
ax[1].plot(angleDegree,volume)
plt.xlim([50, 80])
plt.ylim([.4, .405])

#ax.set_yticks([0,1,2,3])
#ax.set_yticks([0,1,2,3])
```

```python
print('Maximum Volume = ', np.max(volume))
print('Angle of maximum Volume = ', np.argmax(volume))
print('Minimum Angle = ', angleDegree[0])
print('Maximum Angle = ', angleDegree[359])
```
